```python
import pandas as pd
import numpy as np
```

# Carregando conjunto de dados


```python
df = pd.read_csv('agricultural_raw_material.csv')
```

# Explorando o conjunto

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)

df = df.dropna()

df.isnull().sum()
```

```python
df.dtypes
```

```python
lista = ['Wood pulp price % Change', 'Soft sawnwood price % Change','Softlog price % Change', 'Rubber price % Change','Plywood price % Change',
         'Hide price % change', 'Hard sawnwood price % Change', 'Hard log price % Change', 'Fine wool price % Change',
         'Fine wool Price', 'Cotton price % Change', 'Copra price % Change', 'Copra Price','Coarse wool price % Change', 'Coarse wool Price']

df[lista] = df[lista].astype('float')
df.dtypes
```

```python
df.head()
```

```python
df.Month = pd.to_datetime(df.Month.str.upper(), format='%b%y', yearfirst=False)
df = df.set_index('Month')
```

```python
df.head()
```

# AnÃ¡lise ExploratÃ³ria

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9,5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
df.info()
```

Mapa de Calor

```python
raw_data=['Coarse wool Price',
          'Copra Price',
          'Cotton Price',
          'Fine wool Price',
          'Hard log Price',
          'Hard sawnwood Price',
          'Hide Price',
          'Plywood Price',
          'Rubber Price',
          'Softlog Price',
          'Soft sawnwood Price',
          'Wood pulp Price'
 ] #Dados crus

corrmat = df[raw_data].corr() #CorrelaÃ§Ã£o MatemÃ¡tica

fig = plt.figure(figsize=(12,9))

mask = np.triu(np.ones_like(corrmat, dtype=bool))

sns.heatmap(corrmat, vmax=.8, mask=mask, square = True, annot=True)

plt.show()
```

**O mapa de calor demonstra a correlaÃ§Ã£o entre as matÃ©rias primas. Quanto maior valor, maior correlaÃ§Ã£o.**

**Respondendo Perguntas**



1.   Qual a variaÃ§Ã£o normal do preÃ§o de cada matÃ©ria-prima?



```python
changelist = [
    'Coarse wool price % Change',
    'Copra price % Change',
    'Cotton price % Change',
    'Fine wool price % Change',
    'Hard log price % Change',
    'Hard sawnwood price % Change',
    'Hide price % change',
    'Plywood price % Change',
    'Rubber price % Change',
    'Softlog price % Change',
    'Soft sawnwood price % Change',
    'Wood pulp price % Change']

for i in range(len(changelist)):
  plt.figure(figsize=(12,12))
  df[changelist[i]].hist(figsize=(11,9))
  plt.xlabel('% Change')
  plt.ylabel('count')
  plt.legend(changelist[i], loc = 'upper center', bbox_to_anchor=(1.2,1))
```

Observa-se que a maioria das matÃ©rias-primas tem variaÃ§Ã£o de % frequente ideal inferior a 5.

2. Qual matÃ©ria-prima tem o menor preÃ§o ao longo dos anos?

```python
plt.figure(figsize=(10,10))
listmaterials = ['Coarse wool Price',
          'Copra Price',
          'Cotton Price',
          'Fine wool Price',
          'Hard log Price',
          'Hard sawnwood Price',
          'Hide Price',
          'Plywood Price',
          'Rubber Price',
          'Softlog Price',
          'Soft sawnwood Price',
          'Wood pulp Price']

for i in range(len(listmaterials)):
  plt.subplot(4,3,i+1)
  plt.subplots_adjust(hspace=1,wspace=0.5)
  plt.title(listmaterials[i])
  plt.plot(df[listmaterials[i]])
  plt.xticks(rotation=90)
plt.suptitle("Raw-Materials price comparision")
```

Ã‰ possÃ­vel identificar nos grÃ¡ficos que duas matÃ©rias-prima teve os menores preÃ§os ao longos dos anos: algodÃ£o e borracha. Portanto, Ã© necessÃ¡rio uma comparaÃ§Ã£o entre ambos produtos para concluir qual produto teve o menor preÃ§o no decorrer dos anos.

```python
plt.figure(figsize=(10,10))
plt.plot(df[['Cotton Price', 'Rubber Price']])
plt.xlabel('Years')
plt.ylabel('Prices')
plt.legend(['Cotton Prices', 'Rubber Price'], loc = 'upper center', bbox_to_anchor=(1.2,1))
```

O algodÃ£o Ã© a matÃ©ria-prima de menor valor nos Ãºltimos anos.

RelaÃ§Ã£o de preÃ§os de todas as matÃ©rias-primas

```python
materialslist = [['Coarse wool Price',
          'Copra Price',
          'Cotton Price',
          'Fine wool Price',
          'Hard log Price',
          'Hard sawnwood Price',
          'Hide Price',
          'Plywood Price',
          'Rubber Price',
          'Softlog Price',
          'Soft sawnwood Price',
          'Wood pulp Price']]
plt.figure(figsize=(10,10))
for i in range(len(materialslist)):
  plt.plot(df[materialslist[i]])
  plt.xlabel('Years')
  plt.ylabel('Prices')
  plt.legend(df[materialslist[i]], loc = 'upper center', bbox_to_anchor=(1.2,1))
```

A partir dos grÃ¡ficos pudemos analisar as matÃ©rias-primas em tipos de acordo com seu preÃ§o ao longo dos anos.

Materiais de baixo preÃ§o: algodÃ£o, couro, softlog, log duro, preÃ§o de madeira serrada macia, borracha

Materiais de alto preÃ§o: lÃ£ grossa,copra,lÃ£ fina,madeira serrada dura,pasta de madeira,madeira compensada

3. Quais matÃ©rias-primas teve mudanÃ§a drÃ¡stica de preÃ§o?

```python
lowlist= ['Cotton Price','Hide Price','Softlog Price','Hard log Price','Soft sawnwood Price','Rubber Price']
plt.figure(figsize=(12,12))
plt.xlabel('Years')
plt.ylabel('Prices')

for i in range(len(lowlist)):
  sns.scatterplot(y=df[lowlist[i]], x=df.index)
  plt.legend(lowlist, loc='upper center', bbox_to_anchor=(1.2,1))
```

A mudanÃ§a de preÃ§o Ã© drÃ¡stica para o preÃ§o de tora dura entre materiais de baixa faixa de preÃ§o e entre materiais de alto preÃ§o, sÃ£o os preÃ§os de lÃ£ fina.
